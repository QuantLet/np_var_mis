```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import statsmodels.nonparametric.api as nparam
from statsmodels.nonparametric.kernel_regression import KernelReg
```

```python
df = pd.read_csv('moscow_apartment_listings.csv')
```

```python
df = df[['price','lat','lon','footage','subway_dist_to_center']]
```

```python
df = df[['price','footage']]
```

```python
df = df.apply(lambda t: np.log(t))
```

```python
df
```

```python
df = df.sort_values(by=['footage'])
```

```python
df = df.sample(n=1000,random_state=1)
```

```python
len(df)
```

```python
df.hist()
```

```python
fig, ax = plt.subplots(1, 1,figsize=(20, 10))

plt.plot(df['footage'],df['price'],'o')
```

```python
model = KernelReg(endog=[df.price],exog=[df.footage],reg_type='ll',var_type='c', bw='cv_ls')

```

```python
fig, ax = plt.subplots(1, 1,figsize=(20, 10))

plt.plot(df['footage'],df['price'],'o')

```
